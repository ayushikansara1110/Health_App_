```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('./Datasets/survey lung cancer.csv')
print('DataFrame Shape: ', df.shape)
print(display(df))
```

```python
df.info()
```

```python
map1 = {'F': 0, 'M':1}
map2 = {'NO': 0, 'YES':1}

df['GENDER'] = df['GENDER'].map(map1)
df['LUNG_CANCER'] = df['LUNG_CANCER'].map(map2)
```

```python
df
```

```python
plt.figure(figsize=(16,12))
rows = 5; columns = 3

explode = (0.01, 0.01)


for index, col in enumerate([col for col in df.columns if col !='AGE']):
    plt.subplot(rows,columns,index+1)
    labels = df[col].unique()
    plt.pie(df[col].value_counts(), labels=labels, explode=explode, autopct='%.1f%%')
    # plt.legend()
    plt.title(f'Distribution of {col}') 

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Distribution of Age')
sns.countplot(x=df['AGE'], palette='husl')
# sns.color_palette("crest", as_cmap=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, df['LUNG_CANCER'], test_size=0.2, random_state=42)
```

```python
from sklearn import svm
model = svm.SVC()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
pred = model.predict(X_test)
acc = accuracy_score(pred, y_test)
print("Accuracy:",acc*100)
```

```python
import pickle

with open("lung_cancer.pkl", 'wb') as file:
    pickle.dump(model, file)
```
